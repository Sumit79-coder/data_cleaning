```python
import pandas as pd
```

```python
data = pd.read_excel("/Users/sumitkumargupta/Desktop/Dal /data/two_row.xlsx")
```

```python
data
```

```python
data.fillna('N.A.', inplace=True)
```

```python
data
```

```python
drop_extra = data.drop(columns =["Unnamed: 2"])
drop_extra
```

```python
drop_nans = drop_extra[drop_extra["Species"].notna()]
drop_nans
```

```python
column_values = []
counter = 0
for i in range(0,414):
    if i%2 == 0:
        counter += 1
    column_values.append(counter)
column_values
```

```python
iter_cols = drop_nans
iter_cols["Iteration"] = column_values
iter_cols
```

```python
iter_cols.tail(30)
```

```python
date_dataframe = iter_cols.iloc[1::2]
date_dataframe.drop(columns=["Species", "Volume", "Factor", "n. Counted", "No. cells / ml "], inplace = True)
date_dataframe.rename(columns={ date_dataframe.columns[0]: "Date" }, inplace=True)
date_dataframe
```

```python
location_dataframe = iter_cols.iloc[0::2]
location_dataframe.drop(columns=["Species", "Volume", "Factor", "n. Counted", "No. cells / ml "], inplace = True)
location_dataframe.rename(columns={ date_dataframe.columns[0]: "Location" }, inplace=True)
location_dataframe
```

```python
final = pd.merge(left=iter_cols,right=date_dataframe, how = "inner", on="Iteration")
final
```

```python
final_1 = pd.merge(left=final,right=location_dataframe, how = "inner", on="Iteration")
final_1
```
